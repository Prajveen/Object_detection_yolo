```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.image as img
```

```python
def dump_images(df):
  for i, row in df.iterrows():
    image = cv2.imdecode(np.frombuffer(row['image'], np.uint8), cv2.IMREAD_COLOR)
    cv2.imwrite(f'./dataset/{i}.jpg', image)
    
```

```python
easy_images_df = pd.read_parquet('easy-500/images.parquet')
easy_labels_df = pd.read_parquet('easy-500/labels.parquet')

hard_images_df = pd.read_parquet('hard-500/images.parquet')
hard_labels_df = pd.read_parquet('hard-500/labels.parquet')

easy_images = dump_images(easy_images_df)
hard_images = dump_images(hard_images_df)
```
